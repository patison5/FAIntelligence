```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

from matplotlib.pyplot import figure
```

```python
days = [20, 15, 17, 12, 19, 10, 18, 20, 13, 18]
experiments = [9.2, 1.4]

N = 20
```

```python
all_1 = np.array([])
all_2 = np.array([])

sr_1 = []
sr_2 = []

for i, j in enumerate(days):
    for m in experiments:                
        if (m == 9.2):
            users_times = np.ones(j) * m
            random_values = np.random.normal(0, 1, j)
            res = users_times + random_values
            all_1 = np.append( all_1 , res )
            sr_1.append(sum(res) / len(res))
        else:
            users_times = np.ones(j) * m
            random_values = np.random.normal(0, 0.2, j)
            res = users_times + random_values
            all_2 = np.append( all_2 , res )
            sr_2.append(sum(res) / len(res))
        

print(sr_1)
print("")
print(sr_2)

import random

x = np.arange(0,10,1)

fig,ax = plt.subplots()

y_mean_1 = [np.mean(sr_1)]*len(sr_1)
data_line = ax.plot(x, sr_1, label='Время', marker='o')
mean_line = ax.plot(x, y_mean_1, label='Среднее', linestyle='--')

legend = ax.legend(loc='upper right')

plt.figure(figsize=(10,8))
plt.show()
```

```python
users = np.array([5, 12, 9, 4, 8, 11, 5, 11, 6, 4])
delta = np.random.randint(3, 10, size=len(users))
data = []
data_users = []

for i, j in enumerate(users):
    data_users.append(users[i])
    data.append(users[i] + delta[i])
    
print(data_users)
print(data)

x = np.arange(0, len(data), 1)

fig, ax = plt.subplots()


data_line = ax.plot(x, data, label='Общее кол-во', marker='o')
mean_line = ax.plot(x, users, label='Использовавших', marker='o')

legend = ax.legend(loc='upper right')

plt.figure(figsize=(10,8))
plt.show()



# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
IT = data
ECE = data_users
 
# Set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
 
# Make the plot
plt.bar(br1, IT, color ='r', width = barWidth,
        edgecolor ='grey', label ='Общее кол-во')
plt.bar(br2, ECE, color ='g', width = barWidth,
        edgecolor ='grey', label ='Использовавших ')
 
# Adding Xticks
plt.xlabel('Дни', fontweight ='bold', fontsize = 15)
plt.ylabel('Кол-во пользователей', fontweight ='bold', fontsize = 15)

 
plt.legend()
plt.show()
```

```python
random_values = np.ones(15) * 3
random_normals = np.random.normal(0, 0.4, 15)
res = random_values + random_normals
delta = np.random.uniform(0, 6, size=len(random_values))


print(res)


values = []
delta_values = []

for i, j in enumerate(res):
    values.append(res[i])
    delta_values.append(res[i] + delta[i])
    
print(delta_values)


# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
IT = delta_values
ECE = values
 
# Set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
 
# Make the plot
plt.bar(br1, IT, color ='r', width = barWidth, label ='До')
plt.bar(br2, ECE, color ='g', width = barWidth, label ='После ')
 
# Adding Xticks
plt.xlabel('Дни тестирования', fontweight ='bold', fontsize = 15)
plt.ylabel('Затраченное время', fontweight ='bold', fontsize = 15)

 
plt.legend()
plt.show()
```
