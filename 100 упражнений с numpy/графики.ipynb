```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

FA = [1729, 5565]
FB = [156137, 183678]

br1 = np.arange(len(FA))
br2 = [x + barWidth for x in br1]

plt.bar(br1, FB, color ='r', width = barWidth,
        edgecolor ='grey', label ='Размер файла до очистки (Число символов)')
plt.bar(br2, FA, color ='g', width = barWidth,
        edgecolor ='grey', label ='Размер файла после очистки (Число символов)')

# Adding Xticks
plt.xlabel('Набор данных', fontweight ='bold', fontsize = 15)
plt.ylabel('Кол-во символов', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(FA))],
        ['1', '2'])

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]
 
# Set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
 
# Make the plot
plt.bar(br1, IT, color ='r', width = barWidth,
        edgecolor ='grey', label ='IT')
plt.bar(br2, ECE, color ='g', width = barWidth,
        edgecolor ='grey', label ='ECE')
plt.bar(br3, CSE, color ='b', width = barWidth,
        edgecolor ='grey', label ='CSE')
 
# Adding Xticks
plt.xlabel('Branch', fontweight ='bold', fontsize = 15)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IT))],
        ['2015', '2016', '2017', '2018', '2019'])
 
plt.legend()
plt.show()
```

```python
after = [1.0, 5.0,1.0, 5.0,1.0, 5.0, 1.0, 5.0,1.0, 5.0,1.0, 5.0]
before = [152.0, 179.0,152.0, 179.0,152.0, 179.0,152.0, 179.0,152.0, 179.0,152.0, 179.0]

N = len(after)

ind = np.arange(N)  
width = 0.9
 
fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, after, width)
p2 = plt.bar(ind, before, width,
             bottom = after)
 
plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind, ("k_" + str(f) for f in range(0, len(after))))
plt.yticks(np.arange(0, max(before), 10))
plt.legend((p1[0], p2[0]), ('Размер файла до обработки', 'Размер файла после обработки'))
 
plt.show()
```
